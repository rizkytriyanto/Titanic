```python
pd_train = pd.read_csv('train.csv')
Alone = [0 for k in range(len(pd_train))]
for p in range(len(pd_train)):
    if pd_train['SibSp'][p] == 0 and pd_train['Parch'][p] == 0:
        Alone[p] = 1

pd_train = pd_train.assign(IsAlone =Alone)
pd_train['IsAlone'].value_counts()

fig, ax = plt.subplots(1, 1, figsize  =(15,10))
sns.swarmplot(x='Sex', y='Age', data=pd_train, hue='IsAlone', palette='Set1')

insight = '''
Kebanyakan pria berpergian sendiri
dan berumur antara 20-40 tahun, 
sedangkan wanita kebanyakan tidak
berpergian sendiri dan 
berumur diantara 20-30 tahun
'''

ax.text(0.5,85,'Persebaran Orang yang Bepergian Sendiri',horizontalalignment='center',color='cornflowerblue',fontsize=24,fontweight='bold')
ax.text(0.2,40,insight,horizontalalignment='left',color='grey',fontsize=16,fontweight='normal')
ax.set_xlabel('Jenis Kelamin', fontsize=16)
ax.set_ylabel('Umur', fontsize=16);
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_boston

boston = load_boston()
print(boston.data.shape)

df_boston = pd.DataFrame(data = np.c_[boston['data'], boston['target']],
                        columns = list(boston['feature_names'])+['MEDV'])
df_boston.head()
```

```python
df_boston.describe()
```

```python
plt.figure(figsize = (15,10))
sns.heatmap(df_boston.corr(), annot = True)
```

# REGRESI LINEAR SEDERHANA

```python
x = df_boston['RM'].values.reshape(-1,1)
```

```python
y = df_boston['MEDV'].values
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(x,y)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
print ('harga rumah yang memiliki 12 kamar adalah : {:.2f} $ '. format(float(model.predict(np.array([[12]])))))
```

```python
plt.figure(figsize=(12,10));
sns.regplot(x,y);
plt.xlabel('average number of rooms per dwelling')
plt.ylabel(' Median value of owner-occupied homes in \$1000')
plt.show()
```

```python
sns.jointplot(x='RM', y='MEDV', data= df_boston, kind='reg', size=10);
plt.show()
```

# Evaluasi Regresi

```python
from sklearn.model_selection import train_test_split

x = df_boston.iloc[:, :-1].values
y = df_boston['MEDV'].values

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=0)

lr = LinearRegression()
lr.fit(x_train, y_train)

y_train_pred = lr.predict(x_train)
y_test_pred = lr.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(y_train, y_train_pred)
```

```python
mean_squared_error(y_test, y_test_pred)
```

# Coefficient of Determination

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_train, y_train_pred)
```

```python
r2_score(y_test, y_test_pred)
```

```python
df_train = pd.read_csv('train.csv')
df_train = df_train.fillna(df_train.mean())
print (df_train.shape)
df_train.head()
```

```python
X = df_train['Age'].values.reshape(-1,1)
```

```python
y = df_train['Survived'].values
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logres = LogisticRegression()
```

```python
logres.fit(X, y)
```

```python
print ('0 berarti TIDAK SELAMAT. 0 berarti SELAMAT' , '\n')
print ('prediksi orang berumur 50 tahun: {}' .format(str(logres.predict(np.array([[50]])))))
```

```python
print ('propabilitas orang yang berumur 50 tahun: ', '\n')
print ('probabilitas selamat {:.2f}', format(float(logres.predict_proba(np.array([[50]]))[0,0])))
print ('probabilitas tak selamat {:.2f}', format(float(logres.predict_proba(np.array([[50]]))[0,1])))
```
